# 修剪影片長度

```python
import dlib
import cv2

# 原影片(超過一分鐘版)
input_video_name = './original.mp4'

# 輸出影片(一分鐘版)
new_video_name = './edit.mp4'

# 開啟影片檔案
cap = cv2.VideoCapture(input_video_name)

# 取得畫面尺寸
w = int(cap.get(cv2.CAP_PROP_FRAME_WIDTH))           
h = int(cap.get(cv2.CAP_PROP_FRAME_HEIGHT))
fps = cap.get(cv2.CAP_PROP_FPS)

# 使用 XVID 編碼器建立 VideoWriter 物件
fourcc = cv2.VideoWriter_fourcc(*'H264')

# 設定要處理的時間區間，單位為毫秒
start_time = 0
end_time = 63

# 設定影片播放位置到指定的時間區間開始時間
cap.set(cv2.CAP_PROP_POS_MSEC, start_time*1000)

# 輸出影片，FPS 值為原始影片的 FPS 值
out = cv2.VideoWriter(new_video_name, fourcc, fps, (w, h))     

# 以迴圈從影片檔案讀取影格，並顯示出來
while(cap.isOpened() and cap.get(cv2.CAP_PROP_POS_MSEC)) <= end_time*1000:
    ret, frame = cap.read()
    if not ret or cv2.waitKey(1) & 0xFF == 27 : break
    
    # 將處理結果寫入影片
    out.write(frame)

cap.release()
out.release()
cv2.destroyAllWindows()
cv2.waitKey(1)
```

# 1. Rotation

```python
import cv2

# 開啟影片檔案
cap = cv2.VideoCapture(input_video_name)

# 取得畫面尺寸
w = int(cap.get(cv2.CAP_PROP_FRAME_WIDTH))           
h = int(cap.get(cv2.CAP_PROP_FRAME_HEIGHT))
fps = cap.get(cv2.CAP_PROP_FPS)

# 使用 XVID 編碼器建立 VideoWriter 物件
fourcc = cv2.VideoWriter_fourcc(*'XVID')

# 定義特效
effect_name = 'rotate'

# 設定要處理的時間區間，單位為毫秒
start_time = start_times[effect_name]
end_time = end_times[effect_name]

# 設定影片播放位置到指定的時間區間開始時間
cap.set(cv2.CAP_PROP_POS_MSEC, start_time*1000)

# 輸出影片，FPS 值為原始影片的 FPS 值
out = cv2.VideoWriter('./1.mp4', fourcc, fps, (w, h))

# 以迴圈從影片檔案讀取影格，並顯示出來
while cap.isOpened() and cap.get(cv2.CAP_PROP_POS_MSEC) <= end_time*1000:
    ret, frame = cap.read()
    if not ret:
        break
    
    # 計算當前幀應該旋轉的角度和縮放比例
    progress = (cap.get(cv2.CAP_PROP_POS_MSEC) - start_time*1000) / (end_time*1000 - start_time*1000)
    angle = progress * 360
    scale = progress + 0.1  # 由小到大，每一帧增加 0.1 的缩放比例

    # 旋轉並縮放當前幀
    center = (w // 2, h // 2)
    matrix = cv2.getRotationMatrix2D(center, angle, scale)
    frame = cv2.warpAffine(frame, matrix, (w, h)) 
            
    # 在影片新增文字(標題)
    cv2.putText(frame, text_list[effect_name], (20, 80), topic_font, topic_font_scale, topic_color, topic_thickness, cv2.LINE_AA)
            
    # 將處理結果寫入影片
    out.write(frame)
    
    # # 顯示畫面
    # cv2.imshow("frame", frame)

cap.release()
out.release()
cv2.destroyAllWindows()
cv2.waitKey(1)
```

```python
import numpy as np
import cv2

# 開啟影片檔案
cap = cv2.VideoCapture(input_video_name)

# 取得畫面尺寸
w = int(cap.get(cv2.CAP_PROP_FRAME_WIDTH))           
h = int(cap.get(cv2.CAP_PROP_FRAME_HEIGHT))
fps = cap.get(cv2.CAP_PROP_FPS)

# 使用 XVID 編碼器建立 VideoWriter 物件
fourcc = cv2.VideoWriter_fourcc(*'H264')

# 定義特效
effect_name = 'gray_deepen'

# 設定要處理的時間區間，單位為毫秒
start_time = start_times[effect_name]
end_time = end_times[effect_name]

# 設定影片播放位置到指定的時間區間開始時間
cap.set(cv2.CAP_PROP_POS_MSEC, start_time*1000)

# 輸出影片，FPS 值為原始影片的 FPS 值
out = cv2.VideoWriter('./5.mp4', fourcc, fps, (w, h))

# 設定灰度值初始值
gray_level = 0

# 以迴圈從影片檔案讀取影格，並顯示出來
while(cap.isOpened() and cap.get(cv2.CAP_PROP_POS_MSEC)) <= end_time*1000:
    ret, frame = cap.read()
    if not ret or cv2.waitKey(1) & 0xFF == 27 : break
    
    # 將彩色影像轉換為灰度影像
    gray = cv2.cvtColor(frame, cv2.COLOR_BGR2GRAY)
    
    # 調整灰度值深度
    gray_level = min(gray_level + 1, 255)
    frame = cv2.addWeighted(gray, 1, np.zeros(gray.shape, gray.dtype), 0, gray_level)
            
    # 將處理結果寫入影片
    out.write(frame)
    
    # # 顯示畫面
    # cv2.imshow("frame", frame)
        
cap.release()
out.release()
cv2.destroyAllWindows()
cv2.waitKey(1)
```

# 10. 4 in 1 (Sobel + Scharr + Laplacian + Canny)

```python
import cv2
import numpy as np

# 開啟影片檔案
cap = cv2.VideoCapture(input_video_name)

# 取得畫面尺寸
w = int(cap.get(cv2.CAP_PROP_FRAME_WIDTH))           
h = int(cap.get(cv2.CAP_PROP_FRAME_HEIGHT))
fps = cap.get(cv2.CAP_PROP_FPS)

# 設定縮小後的尺寸
new_w = int(w / 2)
new_h = int(h / 2)

# 使用 XVID 編碼器建立 VideoWriter 物件
fourcc = cv2.VideoWriter_fourcc(*'H264')

# 定義特效
effect_name = 'sobel'

# 設定要處理的時間區間，單位為毫秒
start_time = start_times[effect_name]
end_time = end_times[effect_name]

# 設定影片播放位置到指定的時間區間開始時間
cap.set(cv2.CAP_PROP_POS_MSEC, start_time*1000)

# 輸出影片，FPS 值為原始影片的 FPS 值
out = cv2.VideoWriter('./10.mp4', fourcc, fps, (w, h))

# 以迴圈從影片檔案讀取影格，並顯示出來
while(cap.isOpened() and cap.get(cv2.CAP_PROP_POS_MSEC)) <= end_time*1000:
    ret, frame = cap.read()
    if not ret or cv2.waitKey(1) & 0xFF == 27 : break

    # 縮小影片
    small_frame = cv2.resize(frame, (new_w, new_h))

    # 將4個縮小後的影片組合到同一視窗中
    top_left = small_frame                     # 左上
    top_left = cv2.cvtColor(top_left, cv2.COLOR_BGR2GRAY)
    sobel_x = cv2.Sobel(top_left, cv2.CV_64F, 1, 0, ksize=5)
    sobel_x = cv2.convertScaleAbs(sobel_x)
    top_left = cv2.cvtColor(sobel_x, cv2.COLOR_GRAY2BGR)
    
    top_right = small_frame                    # 右上    
    middle_gray = cv2.cvtColor(top_right, cv2.COLOR_BGR2GRAY)
    canny = cv2.Canny(middle_gray, 100, 200)
    top_right = cv2.cvtColor(canny, cv2.COLOR_GRAY2BGR)
    
    bottom_left = small_frame                  # 左下
    right_gray = cv2.cvtColor(bottom_left, cv2.COLOR_BGR2GRAY)
    laplacian = cv2.Laplacian(right_gray, cv2.CV_64F)
    laplacian = cv2.convertScaleAbs(laplacian)
    bottom_left = cv2.cvtColor(laplacian, cv2.COLOR_GRAY2BGR)

    bottom_right = small_frame                 # 右下
    # 進行Sobel邊緣檢測
    scharrx = cv2.Scharr(bottom_right, cv2.CV_64F, 1, 0)   
    scharry = cv2.Scharr(bottom_right, cv2.CV_64F, 0, 1)
    
    scharrx = cv2.convertScaleAbs(scharrx)
    scharry = cv2.convertScaleAbs(scharry)
    # 將 x 和 y 方向的邊緣檢測結果合併
    bottom_right = cv2.addWeighted(scharrx, 0.5, scharry, 0.5, 0) 
    
    top_row = np.hstack((top_left, top_right))
    bottom_row = np.hstack((bottom_left, bottom_right))
    combined = np.vstack((top_row, bottom_row))

    # 增加第1行文字
    cv2.putText(combined, 'Sobel', (740, 500), font, font_scale, color, thickness, cv2.LINE_AA)
    # 增加第2行文字
    cv2.putText(combined, 'Laplacian', (1580, 500), font, font_scale, color, thickness, cv2.LINE_AA)
    # 增加第3行文字
    cv2.putText(combined, 'Canny', (720, 1040), font, font_scale, color, thickness, cv2.LINE_AA)
    # 增加第4行文字
    cv2.putText(combined, 'Scharr', (1670, 1040), font, font_scale, color, thickness, cv2.LINE_AA)
    
    # 在影片新增文字(標題)
    cv2.putText(combined, text_list[effect_name], (20, 80), topic_font, topic_font_scale, topic_color, topic_thickness, cv2.LINE_AA)

    out.write(combined)
    
    # # 顯示畫面
    # cv2.imshow("frame", combined)

cap.release()
out.release()
cv2.destroyAllWindows()
cv2.waitKey(1)
```

```python
import cv2
import numpy as np

# 開啟影片檔案
cap = cv2.VideoCapture(input_video_name)

# 取得畫面尺寸
w = int(cap.get(cv2.CAP_PROP_FRAME_WIDTH))           
h = int(cap.get(cv2.CAP_PROP_FRAME_HEIGHT))
fps = cap.get(cv2.CAP_PROP_FPS)

# 設定縮小後的尺寸
new_w = int(w / 2)
new_h = int(h / 2)

# 使用 XVID 編碼器建立 VideoWriter 物件
fourcc = cv2.VideoWriter_fourcc(*'H264')

# 定義特效
effect_name = 'gamma'

# 設定要處理的時間區間，單位為毫秒
start_time = start_times[effect_name]
end_time = end_times[effect_name]

# 設定影片播放位置到指定的時間區間開始時間
cap.set(cv2.CAP_PROP_POS_MSEC, start_time*1000)

# 輸出影片，FPS 值為原始影片的 FPS 值
out = cv2.VideoWriter('./17.mp4', fourcc, fps, (w, h))

# 設定gamma值
gamma1 = 0.5
gamma2 = 0.75
gamma3 = 1.25
gamma4 = 1.5

# 建立Gamma轉換表
table1 = np.array([((i / 255.0) ** gamma1) * 255 for i in np.arange(0, 256)]).astype("uint8")
table2 = np.array([((i / 255.0) ** gamma2) * 255 for i in np.arange(0, 256)]).astype("uint8")
table3 = np.array([((i / 255.0) ** gamma3) * 255 for i in np.arange(0, 256)]).astype("uint8")
table4 = np.array([((i / 255.0) ** gamma4) * 255 for i in np.arange(0, 256)]).astype("uint8")

# 以迴圈從影片檔案讀取影格，並顯示出來
while(cap.isOpened() and cap.get(cv2.CAP_PROP_POS_MSEC)) <= end_time*1000:
    ret, frame = cap.read()
    if not ret or cv2.waitKey(1) & 0xFF == 27 : break

    # 縮小影片
    small_frame = cv2.resize(frame, (new_w, new_h))

    # 將4個縮小後的影片組合到同一視窗中
    top_left = small_frame                     # 左上
    top_left = cv2.LUT(top_left, table1)

    top_right = small_frame                    # 右上
    top_right = cv2.LUT(top_right, table2)
    
    bottom_left = small_frame                  # 左下
    bottom_left = cv2.LUT(bottom_left, table3)

    bottom_right = small_frame                 # 右下
    bottom_right = cv2.LUT(bottom_right, table4)

    top_row = np.hstack((top_left, top_right))
    bottom_row = np.hstack((bottom_left, bottom_right))
    combined = np.vstack((top_row, bottom_row))

    # 增加文字
    cv2.putText(combined, 'Gamma : 0.50', (450, 500), font, font_scale, color, thickness, cv2.LINE_AA)
    cv2.putText(combined, 'Gamma : 0.75', (1410, 500), font, font_scale, color, thickness, cv2.LINE_AA)
    cv2.putText(combined, 'Gamma : 1.25', (450, 1040), font, font_scale, color, thickness, cv2.LINE_AA)
    cv2.putText(combined, 'Gamma : 1.50', (1410, 1040), font, font_scale, color, thickness, cv2.LINE_AA)
    
    # 在影片新增文字(標題)
    cv2.putText(combined, text_list[effect_name], (20, 80), topic_font, topic_font_scale, topic_color, topic_thickness, cv2.LINE_AA)

    out.write(combined)
    
    # # 顯示畫面
    # cv2.imshow("frame", combined)

cap.release()
out.release()
cv2.destroyAllWindows()
cv2.waitKey(1)
```

```python
import numpy as np
import cv2

# 開啟影片檔案
cap = cv2.VideoCapture(input_video_name)

# 取得畫面尺寸
w = int(cap.get(cv2.CAP_PROP_FRAME_WIDTH))           
h = int(cap.get(cv2.CAP_PROP_FRAME_HEIGHT))
fps = cap.get(cv2.CAP_PROP_FPS)

# 使用 XVID 編碼器建立 VideoWriter 物件
fourcc = cv2.VideoWriter_fourcc(*'XVID')

# 定義特效
effect_name = 'mask'

# 設定要處理的時間區間，單位為毫秒
start_time = start_times[effect_name]
end_time = end_times[effect_name]

# 設定影片播放位置到指定的時間區間開始時間
cap.set(cv2.CAP_PROP_POS_MSEC, start_time*1000)

# 輸出影片，FPS 值為原始影片的 FPS 值
out = cv2.VideoWriter('./21.mp4', fourcc, fps, (w, h))

# 設定矩形遮罩的初始大小
rect_size = (w, h)

# 以迴圈從影片檔案讀取影格，並顯示出來
while(cap.isOpened() and cap.get(cv2.CAP_PROP_POS_MSEC)) < end_time*1000:
    ret, frame = cap.read()
    if not ret or cv2.waitKey(1) & 0xFF == 27 : break
    
    # 取得目前影格的時間
    current_time = cap.get(cv2.CAP_PROP_POS_MSEC) / 1000.0

    if current_time >= start_time and current_time < end_time:
        # 建立矩形遮罩
        mask = np.zeros_like(frame)
        cv2.rectangle(
            img=mask,
            pt1=(int((w - rect_size[0]) / 2), int((h - rect_size[1]) / 2)),
            pt2=(int((w + rect_size[0]) / 2), int((h + rect_size[1]) / 2)),
            color=(255, 255, 255),
            thickness=-1
        )

        # 合併黑罩和影像
        frame = cv2.bitwise_and(frame, mask)

        # 縮小矩形遮罩的大小
        if rect_size[0] > 0 and rect_size[1] > 0:
            rect_size = (max(rect_size[0] - 10, 1), max(rect_size[1] - 10, 1))

    # 在影片新增文字(標題)
    cv2.putText(frame, text_list[effect_name], (20, 80), topic_font, topic_font_scale, topic_color, topic_thickness, cv2.LINE_AA)
    
    # 寫入處理後的影格到新的影片檔案中
    out.write(frame)
    
    # # 顯示畫面
    # cv2.imshow("frame", frame)

    # 縮小矩形遮罩的大小
    rect_size = (max(rect_size[0] - 10, 1), max(rect_size[1] - 10, 1))

cap.release()
out.release()
cv2.destroyAllWindows()
cv2.waitKey(1)
```

# 合併影片

```python
import cv2
import numpy as np

# 開啟影片檔案
cap = cv2.VideoCapture(input_video_name)

# 取得畫面尺寸
w = int(cap.get(cv2.CAP_PROP_FRAME_WIDTH))           
h = int(cap.get(cv2.CAP_PROP_FRAME_HEIGHT))
fps = cap.get(cv2.CAP_PROP_FPS)
total_duration = 63


new_video_name = './0-1.合併.mp4'

# 建立新的影片檔案
fourcc = cv2.VideoWriter_fourcc(*'mp4v')
video_writer = cv2.VideoWriter(new_video_name, fourcc, fps, (w, h))

# 欲讀取的影片檔案清單
video_files = ['1.mp4', '2.mp4', '3.mp4',
               '4.1.mp4', '5.1.mp4', '6.1.mp4',
               '7.mp4', '8.mp4', '9.mp4',
               '10.mp4', '11.mp4', '12.mp4',
               '13.mp4', '14.mp4', '15.mp4',
               '16.mp4', '17.mp4', '18.mp4',
               '19.mp4', '20.mp4' , '21.mp4']

# 將所有影片檔案讀取進來
video_clips = []
for filename in video_files:
    clip = cv2.VideoCapture(filename)
    video_clips.append(clip)

# 計算每個影片檔案的長度
clip_duration = 3
num_clips = len(video_clips)
clip_lengths = [clip_duration] * num_clips

# 計算每個影片檔案的起始時間
clip_start_times = [sum(clip_lengths[:i]) for i in range(num_clips)]

# 計算每一幀的時間間隔
frame_duration = 1.0 / fps

# 開始合成影片
for t in np.arange(0, total_duration, frame_duration):

    # 計算目前時間落在哪個影片檔案區間內
    clip_index = -1
    for i in range(num_clips):
        if clip_start_times[i] <= t < clip_start_times[i] + clip_lengths[i]:
            clip_index = i
            break

    # 讀取該影片檔案的當前幀
    if clip_index >= 0:
        clip = video_clips[clip_index]
        clip.set(cv2.CAP_PROP_POS_FRAMES, int((t - clip_start_times[clip_index]) * fps))
        success, frame = clip.read()
    else:
        frame = np.zeros((height, width, 3), dtype=np.uint8)

    # 將當前幀寫入新的影片檔案
    video_writer.write(frame)
    
    # # 顯示畫面
    # cv2.imshow("frame", combined)

# 釋放所有影片檔案的資源
for clip in video_clips:
    clip.release()

# 釋放新的影片檔案的資源
video_writer.release()

print('影片合成完成')
```
